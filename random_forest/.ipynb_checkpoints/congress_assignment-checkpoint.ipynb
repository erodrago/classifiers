```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
missing_values = ["?"]
df = pd.read_csv("data/house-votes-84.data", na_values=missing_values, header=None)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns = ["Class Name","handicapped-infants","water-project-cost-sharing","adoption-of-the-budget-resolution",
              "physician-fee-freeze","el-salvador-aid","religious-groups-in-schools","anti-satellite-test-ban",
              "aid-to-nicaraguan-contras","mx-missile","immigration","synfuels-corporation-cutback","education-spending",
              "superfund-right-to-sue","crime","duty-free-exports", "export-administration-act-south-africa"]
```

```python
df.head()
```

```python
df.describe()
```

```python
# number of missing values
df.isna().sum()
```

```python
# Drop column with alot of missing values
df.drop(['export-administration-act-south-africa'], axis = 1) 
```

```python
# Since theres alot of missing values we will replace it with a measure of central tendancy
# Picked mode measure of central tendancy

for column in df.columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
    
df.head()
```

```python
# Seems there are 3 values in some columns yet it contains categorical variables of yes/no, 
# lets show the no of values for each

def plot_all():
    for column in df.columns:
        sns.factorplot(data=df,kind='count',size=3,aspect=3,x=column)
               
plot_all()
```

# Feature selection

```python
# Divide to features and target variables
X = df.drop("Class Name", axis=1)
y = df["Class Name"] 
```

# One hot encoding
Since our columns contain only yes or no values we will use binary encoding

```python
from sklearn.preprocessing import LabelEncoder

def transform(feature):
    '''    
    function to encode all categorical variables
    '''
    le = LabelEncoder()
    X[feature]=le.fit_transform(X[feature])
```

```python
for col in X.columns:
    transform(col)
X.head()
```

# Predicting which Party(Democratic or Republican) 

```python
from sklearn.model_selection import train_test_split

# Dividing to training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Using Decision trees

```python
from sklearn.tree import DecisionTreeClassifier
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

dtree = DecisionTreeClassifier()
# fit the model
dtree.fit(X_train, y_train)

predictions = dtree.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, predictions))
```

# Using Random Forest

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
rfc = RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
rfc.fit(X_train,y_train)

party_prediction = rfc.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Random Accuracy:",metrics.accuracy_score(y_test, party_prediction))
```

# Predicting vote on physicianFeeFreeze

```python
# Divide to features and target variables
X = df.drop("physician-fee-freeze", axis=1)
y = df["physician-fee-freeze"] 
```

```python
for col in X.columns:
    transform(col)
X.head()
```

```python
from sklearn.model_selection import train_test_split

# Dividing to training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

dtree = DecisionTreeClassifier()
# fit the model
dtree.fit(X_train, y_train)

vote_predictions = dtree.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Decision tree Accuracy:",metrics.accuracy_score(y_test, vote_predictions))
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
rfc = RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
rfc.fit(X_train,y_train)

vote_prediction = rfc.predict(X_test)

print("Random Forest Accuracy:",metrics.accuracy_score(y_test, vote_prediction))
```
